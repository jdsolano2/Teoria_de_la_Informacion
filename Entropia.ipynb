DESARROLLAR UN CÓDIGO QUE CALCULE LA ENTROPÍA DE UNA FUENTE CON SÍMBOLOS. VARIAR EL NÚMERO DE SÍMBOLOS DE LA FUENTE, POR LO MENOS CON TRES FUENTES DE SÍMBOLOS DIFERENTES.

EJEMPLO CON UNA FUENTE DE 10 SIMBOLOS: letras desde la A a la J

```python
# Importar las librerias para hacer los calculos y graficos
import math
import matplotlib.pyplot as plt

# Generar la fuente 1 con 10 simbolos y sus probabilidades
fuente1 = {
    "A":0.14,
    "B":0.09,
    "C":0.12,
    "D":0.07,
    "E":0.11,
    "F":0.08,
    "G":0.10,
    "H":0.13,
    "I":0.06,
    "J":0.10
}

#Mostrar la fuente 1 con sus simbolos y probabilidades
print("Fuente 1:")
for simbolo, probabilidad in fuente1.items():
    print(simbolo, "-->", probabilidad)

#Funcion para calcular la suma de todas las probabilidades de cada simbolo
#suma_probabilidades = sum(fuente1.values())
#print("\nLa suma de probabilidades es:", suma_probabilidades)

#Proceso para realizar el calculo de la entropia
entropia = 0

for simbolo, probabilidad in fuente1.items():
    termino = probabilidad * math.log2(probabilidad)
    entropia += termino
    print(f"{simbolo} -> p = {probabilidad} | p*log2(p) = {termino:.6f}")

entropia = -entropia

print("\nLa entropía de la fuente 1 es:", round(entropia, 6), "bits/símbolo")

# Datos para la gráfica
# -------------------------------
simbolos = list(fuente1.keys())
probabilidades = list(fuente1.values())
# Crear gráfica
# -------------------------------
plt.figure(figsize=(12, 6))
barras = plt.bar(simbolos, probabilidades, edgecolor='black')

# Títulos y etiquetas
plt.title("Distribución de Probabilidades de la Fuente F1", fontsize=16, fontweight='bold')
plt.xlabel("Símbolos", fontsize=12)
plt.ylabel("Probabilidad", fontsize=12)

# Cuadrícula
plt.grid(axis='y', linestyle='--', alpha=0.5)

# Mostrar valor encima de cada barra
for barra in barras:
    altura = barra.get_height()
    plt.text(
        barra.get_x() + barra.get_width()/2,
        altura + 0.003,
        f"{altura:.2f}",
        ha='center',
        va='bottom',
        fontsize=10,
        fontweight='bold'
    )

# Mostrar entropía dentro de la gráfica
plt.text(
    7.5, 0.125,
    f"Entropía = {entropia:.4f} bits/símbolo",
    fontsize=11,
    bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4')
)

plt.ylim(0, max(probabilidades) + 0.03)
plt.show()
```

A CONTINUACION SE REALIZO UN CODIGO EN EL CUAL SE CALCULA LA ENTROPIA DE 3 FUENTES DIFERENTES DE SIMBOLOS. F1 = 10 simbolos, F2 = 15 simbolos y F3 = 20 simbolos. Y se realiza la grafica de la distribucion de probabilidad de cada simbolo de la fuente.

```python
import numpy as np
import matplotlib.pyplot as plt

# ==========================================================
# DEBER 1 - BIMESTRE 1
# TEORIA DE LA INFORMACION
# Calculo de la entropia de tres fuentes de simbolos
# ==========================================================

def calcular_entropia(nombre_fuente, simbolos, probabilidades):
    probabilidades = np.array(probabilidades, dtype=float)
    suma_prob = np.sum(probabilidades)

    # Verificar que la suma de probabilidades sea 1
    if not np.isclose(suma_prob, 1.0):
        raise ValueError(f"La suma de probabilidades de {nombre_fuente} es {suma_prob:.4f} y no es igual a 1.")

    # Calculo de la entropia
    entropia = -np.sum(probabilidades * np.log2(probabilidades))

    # Mostrar resultados ordenados
    print("\n" + "="*60)
    print(f"{nombre_fuente.upper()}")
    print("="*60)
    print(f"Número de símbolos: {len(simbolos)}")
    print(f"Suma de probabilidades: {suma_prob:.2f}")
    print("-"*60)
    print(f"{'Símbolo':<12}{'Probabilidad':<15}")
    print("-"*60)

    for simbolo, prob in zip(simbolos, probabilidades):
        print(f"{simbolo:<12}{prob:<15.4f}")

    print("-"*60)
    print(f"Entropía de la fuente: {entropia:.4f} bits/símbolo")
    print("="*60)

    # Grafica
    plt.figure(figsize=(10, 5))
    barras = plt.bar(simbolos, probabilidades)

    plt.xlabel("Símbolos")
    plt.ylabel("Probabilidad")
    plt.title(f"Distribución de Probabilidades - {nombre_fuente}")
    plt.grid(axis='y', linestyle='--', alpha=0.5)

    # Mostrar valor encima de cada barra
    for barra in barras:
        altura = barra.get_height()
        plt.text(
            barra.get_x() + barra.get_width()/2,
            altura + 0.003,
            f"{altura:.2f}",
            ha='center',
            va='bottom',
            fontsize=9
        )

    # Mostrar entropía dentro de la gráfica
    plt.text(
        len(simbolos) / 2,
        max(probabilidades) * 0.85,
        f"Entropía = {entropia:.4f} bits/símbolo",
        fontsize=11,
        bbox=dict(facecolor='white', edgecolor='black', boxstyle='round,pad=0.4')
    )

    plt.ylim(0, max(probabilidades) + 0.05)
    plt.show()

    return entropia


# =========================
# FUENTE 1: 10 SÍMBOLOS
# =========================
simbolos1 = list("ABCDEFGHIJ")
probabilidades1 = [0.18, 0.16, 0.14, 0.12, 0.10, 0.08, 0.07, 0.06, 0.05, 0.04]

# =========================
# FUENTE 2: 15 SÍMBOLOS
# =========================
simbolos2 = list("ABCDEFGHIJKLMNO")
probabilidades2 = [0.14, 0.12, 0.10, 0.09, 0.08, 0.07, 0.06, 0.06, 0.05, 0.05, 0.04, 0.04, 0.04, 0.03, 0.03]

# =========================
# FUENTE 3: 20 SÍMBOLOS
# =========================
simbolos3 = list("ABCDEFGHIJKLMNOPQRST")
probabilidades3 = [0.11, 0.10, 0.09, 0.08, 0.07, 0.07, 0.06, 0.06, 0.05, 0.05, 0.04, 0.04, 0.04, 0.03, 0.03, 0.03, 0.02, 0.01, 0.01, 0.01]

# Calculo de entropias
entropia1 = calcular_entropia("Fuente 1", simbolos1, probabilidades1)
entropia2 = calcular_entropia("Fuente 2", simbolos2, probabilidades2)
entropia3 = calcular_entropia("Fuente 3", simbolos3, probabilidades3)

# Resumen final
print("\n" + "="*60)
print("RESUMEN FINAL")
print("="*60)
print(f"Entropía Fuente 1 (10 símbolos): {entropia1:.4f} bits/símbolo")
print(f"Entropía Fuente 2 (15 símbolos): {entropia2:.4f} bits/símbolo")
print(f"Entropía Fuente 3 (20 símbolos): {entropia3:.4f} bits/símbolo")
print("="*60)
```
